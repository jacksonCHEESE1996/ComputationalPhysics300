```python

import numpy as np
from numba import jit
from datetime import datetime
import matplotlib.pyplot as plt
import random

#a) Write a program that generates and prints out two random numbers between 1 and 6, to simulate the rolling of two dice.


@jit
def dice_roll():
    "simulate roll of 2 dice"
    
    roll_1 = random.randint(1, 6)
    roll_2 = random.randint(1, 6)
    roll = [roll_1, roll_2]
    return roll

print(dice_roll())
```

```python
#b) Modify your program to simulate the rolling of two dice a million times (N0) and count the number of times you get double six (N1). Calculate the probability of finding double six in your simulations (N1/N0)
    


@jit
def double_six(n):
    "simulate 2 dice roll n times"
    "count if double 6s"
    
    double_6_count = 0
    
    for i in list(range(n)):
        
        roll = dice_roll()
        #print(roll)
        if roll[0] & roll[1] == 6:
            double_6_count = double_6_count + 1
        
        else:
            continue
        
    return double_6_count/n

double_six(1000000)
```
