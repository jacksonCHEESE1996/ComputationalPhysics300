```python
# -*- coding: utf-8 -*-
"""
Created on Mon Sep 20 16:05:36 2021

@author: Brian Kerlin

8.5 Homework
"""

import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit


plt.figure(figsize=[10, 6])







def gen_data():
    
    #define the function
    f = lambda x,a,b,c: a*np.exp(-b*x)+c  

    #define the paramters for the plot
    x_min, x_max = 0, 10
    npoints = 100
    a,b,c = 2.5, 1.3, 0.4

    x = np.linspace(x_min, x_max, npoints)
    y_noise = f(x,a,b,c) + np.random.rand(npoints) - 0.5
    y_ideal = f(x,a,b,c)
    
    
    return x, y_noise, y_ideal, x_min, x_max


def dif_poly():
    
    x, y_noise, y_ideal, x_min, x_max = gen_data()
    
    fit_1 = np.polyfit(x, y_noise, 1, full=True)
    fit_2 = np.polyfit(x, y_noise, 2, full=True)
    fit_3 = np.polyfit(x, y_noise, 3, full=True)
    fit_4 = np.polyfit(x, y_noise, 4, full=True)
    fit_5 = np.polyfit(x, y_noise, 5, full=True)
    y_p_1 = np.polyval(fit_1[0], x)
    y_p_2 = np.polyval(fit_2[0], x)
    y_p_3 = np.polyval(fit_3[0], x)
    y_p_4 = np.polyval(fit_4[0], x)
    y_p_5 = np.polyval(fit_5[0], x)

    plt.plot(x,y_ideal, label='ideal function')
    plt.plot(x,y_noise, 'o', label='noisy data')
    plt.plot(x,y_p_1, label='polyfit1')
    plt.plot(x,y_p_2, label='polyfit2')
    plt.plot(x,y_p_3, label='polyfit3')
    plt.plot(x,y_p_4, label='polyfit4')
    plt.plot(x,y_p_5, label='polyfit5')

    plt.xlabel('x')
    plt.ylabel('$f(x)$')
    plt.xlim([x_min,x_max])
    plt.legend(fontsize=12)
    #plt.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
    plt.show()
    #plt.text(2, 2, fit)
  

dif_poly()
```
