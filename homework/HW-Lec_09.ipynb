```python
from numpy import loadtxt
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
data = loadtxt("circular.txt",float)
plt.imshow(data)
#print(data)
plt.savefig("whole_figure.jpeg")

plt.figure()
ft=np.fft.fftn(data)
plt.plot(np.abs(ft.real))
plt.show()
print(np.shape(ft))


plt.figure()
#ft=np.fft.fftn(data)
plt.plot(data[20])
plt.show()
print(np.shape(data[20]))


plt.figure()
ft1=np.fft.fftn(data[20])
b=np.abs(ft1.real)
plt.plot(np.abs(ft1.real))
plt.show()
print(b)

def clean_1(dirty_array_fft):
    "removes the lowest half of coefs"
    da_fft_abs_real = np.abs(dirty_array_fft)
    clean_array = []
    med = np.ma.median(da_fft_abs_real)
    for i in list(range(len(da_fft_abs_real))):
        if da_fft_abs_real[i]>med:
            clean_array.append(dirty_array_fft[i])
        else:
            continue
    return clean_array

g = clean_1(ft1)
print(g)
print(len(b))
print(len(g))
check_1 = np.fft.ifftn(g)
#print(check_1)
plt.figure()
#ft=np.fft.fftn(data)
plt.plot(check_1)
plt.show()


def clean_big(raw_data):
    
    clean_array = []
    
    for i in list(range(len(raw_data))):
        clean_array.append(clean_1(raw_data[i]))
        
    return clean_array
    
data_compressed = clean_big(data)

plt.figure()
plt.imshow(data_compressed, extent=[0,100,0,1], aspect=100)


plt.show()
plt.savefig("whole_figure_compressed.jpeg")




```
